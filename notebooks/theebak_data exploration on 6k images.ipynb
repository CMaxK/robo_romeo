```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../raw_data/text/Flickr8k.token.txt') as token_file:
    lines = token_file.readlines()
```

```python
image_tags = []
captions = []
for line in range(0,len(lines)):
    image_tags.append(lines[line].split('#')[0])
    captions.append(((lines[line].split('#')[1]).split('\t')[1]).split(' .')[0])
```

```python
img_list_file = '../raw_data/text/Flickr_8k.trainImages.txt'
img_list = np.loadtxt(img_list_file, dtype=str)
img_folder_path = '../raw_data/images/'
```

```python
images = 8
rand_img_index = np.random.randint(6001, size=images)
rand_images = [img_list[i] for i in rand_img_index]
fig = plt.figure(figsize=(16,12))
for i in range(0,images):
    plt.subplot(4,int(images/4),i+1)
    img = plt.imread(img_folder_path+rand_images[i])
    plt.imshow(img)
    plt.title(captions[image_tags.index(rand_images[i])])
```
